```
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
nltk.download('punkt')
nltk.download('stopwords')
from collections import defaultdict
import pandas as pd
```

```
path = "/home/jovyan/work/txts/"
files = ["amazon.txt","imdb.txt","yelp.txt"]

stopWords = set(stopwords.words('english'))
stemmer = SnowballStemmer("english")

print(stopWords)
```

```
def Tokenize(line):
    res = []
    tokens = word_tokenize(line)
    for token in tokens:
        t = stemmer.stem(token)
        if t not in stopWords:
            res.append(t)
    return res
```

```
allwords = []
alllines = []
for file in files:
    with open(path+file,"r") as f:
        words = []
        tmp = [] # all lines of file
        for line in f:
            tmp2  = []
            tokens = Tokenize(line)
            for t in tokens:
                if t not in allwords:
                    tmp2.append(t)
                    if t not in words:
                        words.append(t)
            tmp.append(tmp2)
    alllines.append(tmp)
    allwords.append(words)

```

```
def count(wo,li):
    c = 0
    for w in li:
        if w == wo:
            c += 1
    return c
```

```
allanswers = []
for f in alllines:
    filenums = []
    for line in f:
        linenums = []
        for word in allwords[alllines.index(f)]:
            linenums.append(count(word,line))
        filenums.append(linenums)
    allanswers.append(filenums)        
```

```
amazon_df = pd.DataFrame(allanswers[0],columns=allwords[0])
imdb_df = pd.DataFrame(allanswers[1],columns=allwords[1])
yelp_df = pd.DataFrame(allanswers[2],columns=allwords[2])
```
